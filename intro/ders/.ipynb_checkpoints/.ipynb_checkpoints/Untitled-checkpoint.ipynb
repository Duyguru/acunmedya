# Pandas Kütüphanesinden 10 Fonksiyon


```python
import pandas as pd
```

## 1. head() – İlk N Satırı Getirir

```python
df = pd.read_csv("Cartoon_datasets.csv")
df.head(3)
```

## 2. tail() – Son N Satırı Getirir

```python
df.tail(2)
```

## 3. info() – Veri Kümesi Hakkında Bilgi Verir

```python
df.info()
```

## 4. describe() – İstatistiksel Özet Verir

```python
df.describe()
```

## 5. isnull() – Eksik Verileri Kontrol Eder

```python
df.isnull()
```

## 6. drop() – Sütun veya Satır Silme

```python
x=df.drop(0)
x
```

```python
x.head(2)
```

## 7. sort_values() – Verileri Sütuna Göre Sıralama

```python
df.sort_values(by="Rating")
```

## 8. groupby() – Verileri Gruplama

```python
df.groupby("Span")
```

```python
df.groupby("Span").sum()
```

## 9. apply() – Fonksiyon Uygulama

```python
df["Name"].apply(lambda x: x * 2)
```

## 10. merge() – İki DataFrame’i Birleştirme

```python
df2 = df.copy()
merged_df = pd.merge(df, df2, on="Name", suffixes=("_left", "_right"))
merged_df
```
