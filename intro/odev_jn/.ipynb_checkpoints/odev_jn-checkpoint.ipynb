## 1-Boston housing dataseti ML'e hazır hale getirelim.

```python
import pandas as pd
```

```python
df=pd.read_csv("HousingData.csv")
df
```

## 2-Eksik Verileri Dolduralım

```python
df.isnull().sum()
```

```python
df["CRIM"].mean()
```

```python
df["CRIM"]=df["CRIM"].fillna(df["CRIM"].mean())
```

```python
df.isnull().sum()
```

```python
df["ZN"]=df["ZN"].fillna(df["ZN"].mean())
```

```python
df["INDUS"]=df["INDUS"].fillna(df["INDUS"].mean())
```

```python
df["CHAS"]=df["CHAS"].fillna(df["CHAS"].mean())
```

```python
df["AGE"]=df["AGE"].fillna(df["AGE"].mean())
```

```python
df["LSTAT"]=df["ZN"].fillna(df["LSTAT"].mean())
```

```python
df.isnull().sum()
```

## 3-Yeni Dosya Olarak Kaydetme

```python
df.to_csv("HousingData2.csv")
```
